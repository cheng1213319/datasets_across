```python
import pandas as pd
import os

df_processed = pd.read_csv("../dataset_1/database/drug_comb/drugcomb_cleaned_with_mean_int.csv")
df_processed.dropna(subset=['drug_row', 'drug_col', 'cell_line_name'], inplace=True)

df_processed['combo_id'] = df_processed.apply(
    lambda row: f"{row['drug_row']}_{row['drug_col']}_{row['cell_line_name']}", 
    axis=1
)

study_pairs = [('ALMANAC', 'ASTRAZENECA'),
               ('ALMANAC', 'FRIEDMAN'),
               ('ALMANAC', 'ONEIL'),
               ('FRIEDMAN', 'ONEIL')]

output_dir = "study_pair_comparisons"
os.makedirs(output_dir, exist_ok=True)

for study1, study2 in study_pairs:
    study1_data = df_processed[df_processed['study_name'] == study1]
    study2_data = df_processed[df_processed['study_name'] == study2]

    shared_combos = set(study1_data['combo_id']).intersection(set(study2_data['combo_id']))
    
    print(f"{study1},{study2} share {len(shared_combos)}")
    shared_study1 = study1_data[study1_data['combo_id'].isin(shared_combos)]
    shared_study2 = study2_data[study2_data['combo_id'].isin(shared_combos)]
    
    merged = pd.merge(
        shared_study1, 
        shared_study2, 
        on='combo_id', 
        suffixes=(f'_{study1}', f'_{study2}')
    )
    
    # 保存到CSV
    filename = f"{study1}_vs_{study2}.csv"
    merged.to_csv(os.path.join(output_dir, filename), index=False)
    print(f"stored in: {filename}")
```
